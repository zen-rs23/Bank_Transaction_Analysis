```python
import os
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/2023A_36106 Machine Learning Algorithms and Applications/Assignment/Assignment 3/original_dataset'
```

```python
file_list = os.listdir(file_path)
file_list
```

```python
cust_csv = file_list.pop(0)
cust_csv
```

```python
transact_csv = file_list
transact_csv
```

```python
df_append = pd.DataFrame()

for i in transact_csv:
  df_temp = pd.read_csv(file_path+'/'+i, sep = '|')
  df_append = df_append.append(df_temp, ignore_index=True)

df_append
```

```python
df_append.to_csv('mlaa_a3_all_transactions.csv', index=False)
```

```python
cust = pd.read_csv('/content/drive/MyDrive/2023A_36106 Machine Learning Algorithms and Applications/Assignment/Assignment 3/original_dataset/customers.csv', sep='|')
cust.head()
```

```python
pd.set_option('display.max_columns', 100)
mdf = df_append.merge(cust, how='left', on='acct_num')
```
